```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y, linestyle='-.',alpha=0.9,color='r',linewidth=3)
plt.plot(x,z, linestyle=':',alpha=0.9,color='b',linewidth=3)
plt.legend(['sin(x)', 'cos(x)'], loc='best')
plt.xlabel('PI', color='r')
plt.ylabel('sin(x), cos(x)')
plt.grid(color='b',linestyle=':', alpha=0.4)
plt.title('sin(x), cos(x)')
plt.savefig('sin(x),cos(x) graphs.png')
plt.show()
```
