```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y, linestyle='--',alpha=0.9,color='c',linewidth=3)
plt.legend(['sin(x)'], loc='best')
plt.xlabel('PI', color='r')
plt.ylabel('sin(x)',color='b')
plt.grid(color='b',linestyle=':', alpha=0.4)
plt.title('sin(x) graph')
plt.savefig('sin(x) graph.png')
plt.show()
```
