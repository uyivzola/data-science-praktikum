## 3-AMALIYOT. `scatterplot`


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from random import shuffle
%matplotlib inline
```

### Ushbu amaliyotda uybor.uz saytidan yuklab olingan ma'lumotlar tahlili bilan shu'gullanamiz.

```python
colors=['#03071E','#370617','#6A040F','#9D0208','#D00000','#DC2F02','#E85D04','#F48C06','#FAA307','#FFBA08','#6d6875','#9e2a2b']
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

#### Yuqoridagi jadvalda Toshkent shahridagi sotildagian uylar haqida ma'lumotlar jamlangan.

### 1-VAZIFA. Uylarning narxi (`price`) va maydoni (`size`) o'rtasida bog'liqlikni ko'rsatuvchi scatter plot chizing.

```python
#JAVOBINGIZNI SHUYERGA YOZING
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x=df['size'], y=df['price'], hue=df['district'], s=120)
plt.grid(linewidth=0.8, linestyle=':', color='b', alpha=.8)
plt.title('Price with size in Tashkent City | Uzbekistan', fontsize=15)
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel("Price", fontsize=14)
plt.xlabel("Size", fontsize=14)
plt.savefig('uybor - price,size,district all.png')
plt.show()
```

### Jadvalda 7000 dan oshiq uy bor, aksar uylarning maydoni 250kv.m dan kam. Keling noodatiy katta qiymatlarni tashlab yuboramiz va yuqoridagi chizmani qaytadan chizamiz.

```python
#JAVOBINGIZNI SHUYERGA YOZING
df=df[df['size']<250]
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x=df['size'], y=df['price'], hue=df['district'], s=100)
plt.grid(linewidth=0.8, linestyle=':', color='b', alpha=.8)
plt.title('Price with size in Tashkent City | Uzbekistan (less than 250 kv.sq)', fontsize=15)
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel("Price", fontsize=14)
plt.xlabel("Size", fontsize=14)
plt.savefig('uybor  - price, size and district less 250km.sq.png')
plt.show()
```

### 2-VAZIFA. Uylarning narxi (price) va maydoni (size) o'rtasida bog'liqlikni ko'rsatuvchi scatter va to'g'ri chiziqni chizing.

```python
# JAVOB UCHUN JOY
plt.figure(figsize=(20,10))
sns.regplot(data=df, x=df['size'], y=df['price'])
plt.grid(linewidth=0.8, linestyle=':', color='b', alpha=.8)
plt.title('Connection of Homes price with size in Tashkent City | Uzbekistan', fontsize=15)
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel("Price", fontsize=14)
plt.xlabel("Size", fontsize=14)
plt.savefig('uybor - price, size.png')
plt.show()
```

### 3-VAZIFA. Uylarning narxi, maydoni va xonalar soni o'rtasidagi bog'liqlikni ko'rsating.

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x=df['size'], y=df['price'], hue=df['rooms'], palette=colors)
plt.grid(linewidth=0.8, linestyle=':', color='b', alpha=.8)
plt.title('Homes price, room and  size in Tashkent City | Uzbekistan', fontsize=15)
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel("Price", fontsize=14)
plt.xlabel("Size", fontsize=14)
plt.savefig('uybor - price, room, size.png')

plt.show()
```

### 4-VAZIFA. Uylarning narxi, maydoni va tumanlar o'rtasidagi bog'liqlikni ko'rsating.

```python
# JAVOBNI SHUYERGA YOZING
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x=df['size'], y=df['price'], hue=df['district'], s=100)
plt.grid(linewidth=0.8, linestyle=':', color='b', alpha=.8)
plt.title('Homes in Tashkent City | Uzbekistan', fontsize=15)
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel("Price", fontsize=14)
plt.xlabel("Size", fontsize=14)
plt.savefig('uybor - price, district, size.png')
plt.show()
```

### 5-VAZIFA. Swarm ploit yordamida turli tumanlardag uylarning narxini taqqoslang.

```python
# JAVOB UCHUN JOY
plt.figure(figsize=(20,10))
sns.swarmplot(data=df, x=df['district'], y=df['price'],palette=colors)
plt.grid(linewidth=.8, linestyle=':', alpha=.7, color=colors[6])
plt.title('Prices in Districts', fontsize=16, color=colors[7])
plt.yticks(color=colors[7], fontsize=12)
plt.xticks(color=colors[7], fontsize=12, rotation=90)
plt.ylabel('Price', fontsize=14, color=colors[7])
plt.xlabel('Districts', fontsize=14, color=colors[7])
plt.savefig('uybor - swarmplot.png')
plt.show()
```
