```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### 1-VAZIFA. Avvalgi [amaliyotda](https://colab.research.google.com/drive/1jzhJ1gN6msILYOHaG2fhBBvUGaa4tcLf?usp=sharing) saqlangan faylni yangi `df`ga o'qing

```python
#JAVOBINGIZNI SHU YERGA YOZING
df=pd.read_csv('https://github.com/uyivzola/data-science-praktikum/raw/main/module-4%20-%20visualization/weather-report.csv', index_col=0)
df.head()
```

### 2-VAZIFA. `df` barcha ustunlarini chiziqli grafik ko'rinishida chiqaring

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(16,8))
sns.lineplot(data=df, estimator='mean', linewidth=2, marker='o')
plt.title('Toshkent shahri yillik ob-havo ko\'rsatkichlari', fontsize=15)
plt.ylabel('Temperature', fontsize=13)
plt.xlabel('Yil', fontsize=13)
plt.xticks(rotation=90)
plt.savefig('Weather - AllInOne.png')
plt.show()
```

### 3-VAZIFA. Grafiklar nomini to'g'rilang (`T`, `Tm` va hokazo matnlarni o'zbek tilida chiqaring.)

```python
#JAVOBINGIZNI SHU YERGA YOZING
names=pd.read_csv('https://github.com/uyivzola/data-science-praktikum/raw/main/module-4%20-%20visualization/weather-report-accronym.csv', index_col=0)
names.drop('PP', inplace=True)
names
```

```python
df.columns=names['Full Name']
colors=['#03071E','#370617','#6A040F','#9D0208','#D00000','#DC2F02','#E85D04','#F48C06','#FAA307','#FFBA08','#6d6875','#9e2a2b']
plt.figure(figsize=(22,10))
sns.lineplot(data=df, marker='o', linewidth=1.5)
plt.grid(linewidth=.8, linestyle=":", color=colors[5], alpha=0.8)
plt.title('Annual Temprature figures for Tashkent, Uzbekistan', fontsize=15)
plt.legend(fontsize=12, loc='best')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Temperature', fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.savefig('Weather - AllInOne - with legend.png')
plt.show()
```

### 4-VAZIFA. Yuqoridagi grafikda turli ko'rsatkichlar aralashib ketdi. Keling ustun turiga qarab alohida grafiklar chizamiz.

#### 4.1. Haroratni ko'rsatuvchi ustunlarni alohida chiziqli grafik qilib chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
temps=df[['Average annual temperature', 'Annual average maximum temperature',
       'Average annual minimum temperature']]
plt.figure(figsize=(22,10))
sns.lineplot(data=temps, linewidth=1.9, marker='^', markerfacecolor=colors[3])
plt.grid(linewidth=0.9,alpha=0.8, color=colors[5], linestyle=':')
plt.title('Annual Temperature figures for Tashkent, Uzbekistan', fontsize=17)
plt.legend(fontsize=14)
plt.ylabel('Temperature', fontsize=15)
plt.xlabel('Years', fontsize=15)
plt.xticks(rotation=90, fontsize=13)
plt.savefig('Weather - onyl Temperature.png')
plt.show()
```

#### 4.2. Kunlarni ko'rsatuvchi ustunlarni alohida grafik ko'rinishida chiqaring

```python
#JAVOBINGIZNI SHU YERGA YOZING
days=df[['Number of days with rain', 'Number of days with snow',
       'Number of days with storm', 'Number of foggy days',
       'Number of days with tornado', 'Number of days with hail']]
plt.figure(figsize=(22,10))
sns.lineplot(data=days, marker='o', markerfacecolor=colors[7])
plt.grid(linewidth=.8,linestyle=':', color=colors[3], alpha=0.8)
plt.title('Annual weather figures for Tashkent | Uzbekistan'.upper(), fontsize=17, color=colors[5])
plt.legend(fontsize=14)
plt.ylabel('Days'.upper(), fontsize=15, color=colors[7])
plt.xlabel('Year'.upper(), fontsize=15, color=colors[7])
plt.xticks(rotation=90, fontsize=13)
plt.savefig('Weather - Only DAYS.png')
plt.show()
```

### 5-VAZIFA. Yillik o'rtacha shamol tezligini ustunli grafik ko'rinishida chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
wind=df['Annual average wind speed']
plt.figure(figsize=(25,15))
sns.barplot(x=wind.index, y=wind,palette=colors)
plt.grid(color=colors[7], linewidth=.8,alpha=.9, linestyle='-.')
plt.title('Annual weather figures for Tashkent | Uzbekistan'.upper(), fontsize=17, color=colors[5])
plt.ylabel('Speed - km/s'.upper(), fontsize=15, color=colors[7])
plt.xlabel('Year'.upper(), fontsize=15, color=colors[7])
plt.xticks(rotation=90, fontsize=13)
plt.savefig('Weather - only Wind.png')
plt.show()
```

### 6-VAZIFA. Yomg'rili, qorli, bo'ronli, tumanli va do'l yoqqan kunlar sonini heatmap ko'rinishida chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(30,5))
sns.heatmap(days.T,square=True,linewidth=1, cmap='YlGnBu')
plt.title('Weather Reports for Tashkent')
yticks_labels=['Rain','Snow','Foggy','Tornado','Hail']
plt.ylabel('Weather Condition')
plt.xlabel('Year')
plt.yticks(np.arange(5)+.5, labels=yticks_labels, fontsize=15)
plt.xticks(rotation=90, fontsize=15)
plt.savefig('Weather - Heatmap.png')
plt.show()
```
