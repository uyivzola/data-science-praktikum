```python
ratings=pd.read_csv('/kaggle/working/Practicum_Datasets/IMDb/IMDb_ratings.csv')
movies=pd.read_csv('/kaggle/working/Practicum_Datasets/IMDb/IMDb_movies.csv')
```

```python
movies_ratings=pd.merge(movies,ratings)
movies_ratings.head()
```

```python
movies_ratings.columns
```

```python
extracted=movies_ratings[['title','year','genre', 'country','weighted_average_vote']]
extracted.drop(index=83917, inplace=True)
extracted.year.unique()
```

```python
pd.pivot_table(extracted, index=['year', 'country',]).sort_values(by='weighted_average_vote',ascending=False)
```
